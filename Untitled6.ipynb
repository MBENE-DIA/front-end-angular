**Création de 6 figures sur 2 lignes et 3 colonnes**

```python
lignes, colonnes = 2, 3
fig, ax = plt.subplots(lignes, colonnes,
                       sharex='col', 
                       sharey='row')

for row in range(lignes):
    for col in range(colonnes):
        ax[row, col].text(0.5, 0.5, 
                          str((row, col)),
                          color="green",
                          fontsize=18, 
                          ha='center')
plt.show()
```
